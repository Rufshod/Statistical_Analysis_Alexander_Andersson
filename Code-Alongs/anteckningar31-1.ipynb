#### I regressionsanalys antar vi explicit att en variabel kallad responsvariabeln "styrs" av adnra variabler. Kallade förklaringsvariablerna.  
Med regressionsanalys kan vi prediktera (förutsäga) värden hos responsvariabeln givet värden på förklaringsvariablerna.

Man börjar ofta med en scatterplot för att troliggöra ett linjärt samband.  

Om värdet hos responsvariabeln är entydigt bestämt av värdet hos förklaringsvariablerna säger vi att sambandet är deterministiskt.  (känns som att detta aldrig borde hända med riktig data)  
Men om vi som är fallet i verkligheten så är inte sambandet exakt. Då säger vi att sambandet är stokastiskt eller slumpmässigt.  

I regressionsanalys inlkulderar vi en stokastisk felterm för att visa...

$y = \beta_0 + \beta_1x + \epsilon$
där y är responsvariabeln. x är förklaringsvariabeln och epsilon är den stokastiska feltermen. Interceptet (skärningen med y-axeln) \b_0 och lutningen \b_1 är okända parametrar som behöver skattas från observerade data.  
  
När vi använt observerade data för att skatta $\beta_0$ och $\beta_1$ får vi ett deterministiskt samband  
$\hat{y} = \beta_0 + \beta_1x$  
där yhatt är vårt predikterade värde på y givet x och b0 och b1 är skattningar av B0 och B1.  
Prediktionerna yhatt är typiskt inte identiska med observerade värdena på y och vi kallar skillnaderna e = y - yhatt för **residualer**  
  
korrelationskoefficienten tar värden mellan -1 och 1 och är ett mått både på hur starkt ett linjärt samband är och vilken riktning.
