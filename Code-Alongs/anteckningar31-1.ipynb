#### I regressionsanalys antar vi explicit att en variabel kallad responsvariabeln "styrs" av adnra variabler. Kallade frklaringsvariablerna.  
Med regressionsanalys kan vi prediktera (frutsga) vrden hos responsvariabeln givet vrden p frklaringsvariablerna.

Man brjar ofta med en scatterplot fr att troliggra ett linjrt samband.  

Om vrdet hos responsvariabeln r entydigt bestmt av vrdet hos frklaringsvariablerna sger vi att sambandet r deterministiskt.  (knns som att detta aldrig borde hnda med riktig data)  
Men om vi som r fallet i verkligheten s r inte sambandet exakt. D sger vi att sambandet r stokastiskt eller slumpmssigt.  

I regressionsanalys inlkulderar vi en stokastisk felterm fr att visa...

$y = \beta_0 + \beta_1x + \epsilon$
dr y r responsvariabeln. x r frklaringsvariabeln och epsilon r den stokastiska feltermen. Interceptet (skrningen med y-axeln) \b_0 och lutningen \b_1 r oknda parametrar som behver skattas frn observerade data.  
  
Nr vi anvnt observerade data fr att skatta $\beta_0$ och $\beta_1$ fr vi ett deterministiskt samband  
$\hat{y} = \beta_0 + \beta_1x$  
dr yhatt r vrt predikterade vrde p y givet x och b0 och b1 r skattningar av B0 och B1.  
Prediktionerna yhatt r typiskt inte identiska med observerade vrdena p y och vi kallar skillnaderna e = y - yhatt fr **residualer**  
  
korrelationskoefficienten tar vrden mellan -1 och 1 och r ett mtt bde p hur starkt ett linjrt samband r och vilken riktning.
